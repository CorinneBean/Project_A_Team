```python
import pandas as pd
```

```python
file1 = "Demographic Data/demographic_details.csv"
file2 = "../Resources/acc_intake_outcome.csv"
demographics = pd.read_csv(file1)
shelter = pd.read_csv(file2)
```

```python
# Drop columns in demographics DF
demographics.drop(["land_area_in_sqmi", "population_by_age", "school_enrollment_age_3_to_17"], axis = 1, inplace= True)
demographics = demographics.loc[:, ~demographics.columns.str.contains('^Unnamed')]

demographics.head(3)
```

```python
#Rename zipcode_intake to zipcode df.rename(columns={"A": "a", "B": "c"})

shelter.rename(columns = {"zipcode_intake":"zipcode"}, inplace = True)
```

```python
# Drop rows without zip codes
shelter = shelter.dropna(axis=0, subset=['zipcode'])
```

```python
# Merge two DF 

animal_demo_df = shelter.merge(demographics, on= "zipcode")
animal_demo_df.head()
```

```python
#cust_sell = mainDf[mainDf.Type == 'S']

cat_demo = animal_demo_df[animal_demo_df.animal_type_intake == "Cat"]
cat_demo.head()
```

```python
dog_demo = animal_demo_df[animal_demo_df.animal_type_intake == "Dog"]
dog_demo.head()
```

```python
dog_demo.to_csv("../Resources/dog_demo.csv")
cat_demo.to_csv("../Resources/cat_demo.csv")
```
