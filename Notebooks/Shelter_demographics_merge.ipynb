```python
import pandas as pd
from datetime import datetime as dt
import time
```

```python
file1 = "../Resources/Demographics_separated_columns.csv"
file2 = "../Resources/acc_intake_outcome.csv"
demographics = pd.read_csv(file1)
shelter = pd.read_csv(file2)
```

```python
demographics.head()
```

```python
# Drop columns in demographics DF
demographics = demographics.loc[:, ~demographics.columns.str.contains('^Unnamed')]

demographics.head(3)
```

```python
#Rename zipcode_intake to zipcode df.rename(columns={"A": "a", "B": "c"})

shelter.rename(columns = {"zipcode_intake":"zipcode"}, inplace = True)
shelter["zipcode"]= shelter["zipcode"].astype(int)
shelter.head()
```

```python
# Drop rows without zip codes
shelter = shelter.dropna(axis=0, subset=['zipcode'])
```

```python
shelter.columns
```

```python
demographics.columns
```

```python
# Merge two DF 

animal_demo_df = shelter.merge(demographics, how = "left", on= "zipcode")
animal_demo_df.head()
```

```python
animal_demo_df.columns
```

```python
# Convert date/times to datetime
animal_demo_df["datetime_outcome"] = pd.to_datetime(animal_demo_df["datetime_outcome"])
animal_demo_df["date_of_birth_outcome"] = pd.to_datetime(animal_demo_df["date_of_birth_outcome"])
animal_demo_df["datetime_intake"] = pd.to_datetime(animal_demo_df["datetime_intake"])
```

```python
# Convert columns to correct type
# Split Datetime column 
animal_demo_df["outcome_date"] = animal_demo_df["datetime_outcome"].dt.date
animal_demo_df["outcome_year"] = animal_demo_df["datetime_outcome"].dt.year
animal_demo_df["outcome_month"] = animal_demo_df["datetime_outcome"].dt.month
animal_demo_df["outcome_day"] = animal_demo_df["datetime_outcome"].dt.day
animal_demo_df["outcome_time"] = animal_demo_df["datetime_outcome"].dt.time
animal_demo_df["DOB_year"] = animal_demo_df["date_of_birth_outcome"].dt.year
animal_demo_df["DOB_month"]= animal_demo_df["date_of_birth_outcome"].dt.month
animal_demo_df["DOB_day"]= animal_demo_df["date_of_birth_outcome"].dt.day
animal_demo_df["intake_date"] = animal_demo_df["datetime_intake"].dt.date
animal_demo_df["intake_year1"] = animal_demo_df["datetime_intake"].dt.year
animal_demo_df["intake_month1"] = animal_demo_df["datetime_intake"].dt.month
animal_demo_df["intake_day1"] = animal_demo_df["datetime_intake"].dt.day
animal_demo_df["intake_time1"] = animal_demo_df["datetime_intake"].dt.time
```

```python
animal_demo_df.columns
```

```python
# Determine Day of the Week of Outcome
# Convert DateTime into Series so that can extrapolate the day of the week
animal_demo_df["outcome_date"] = pd.to_datetime(animal_demo_df["outcome_date"])
animal_demo_df["intake_date"] = pd.to_datetime(animal_demo_df["intake_date"])
animal_demo_df["intake_date"].dtypes
```

```python
Date_time_series = animal_demo_df.iloc[:,-11]
Date_time_series = Date_time_series.dt.dayofweek
print(f'{animal_demo_df["outcome_date"]} "  "  {animal_demo_df["datetime_outcome"]} " " {Date_time_series}' )
```

```python
animal_demo_df.head()
```

```python
Date_time_series_intake = animal_demo_df.iloc[:,-5]
Date_time_series_intake = Date_time_series_intake.dt.dayofweek
print(f'{animal_demo_df["intake_date"]} "  "  {animal_demo_df["datetime_intake"]} " " {Date_time_series_intake}' )
```

```python
# Add new Day of the Week column to Outcomes_df (0 = Sunday, 1 = Monday, etc)
animal_demo_df["outcome_numeric_weekday"] = Date_time_series
animal_demo_df["intake_numeric_weekday"] = Date_time_series_intake
```

```python
# Generate our categorical variable lists
animal_cat = animal_demo_df.dtypes[animal_demo_df.dtypes == "object"].index.tolist()
```

```python
# Check the number of unique values in each column
animal_demo_df[animal_cat].nunique()
```
