```python
import pandas as pd
```

```python
df = pd.read_csv('../../../../Data Files/AAC_Intake_etl.csv')
df.head(5)
```

```python
location_df = df[['Index_ID','Animal_ID','Found_Location']]
location_df.head(5)
```

```python
location_df['Found_Location_count'] = location_df['Found_Location'].str.count(' ') + 1
```

```python
#location_df=location_df.drop('Found_Location1',axis=1)
```

```python
short_named_citites = location_df[location_df['Found_Location_count'].isin([2,3])]
short_named_citites.Found_Location.unique()


```

```python
short_cities_list = ['Austin (TX)', 'Travis (TX)', 'Del Valle (TX)',
       'Outside Jurisdiction', 'Manor (TX)', 'Pflugerville (TX)',
       'Williamson (TX)', 'Lago Vista (TX)', 'Lakeway (TX)',
       'Cedar Park (TX)', 'Hays (TX)', 'Round Rock (TX)',
       'Jonestown (TX)', 'Leander (TX)', 'Bastrop (TX)', 'Bee Cave (TX)',
       'Creedmoor (TX)', 'Burnet (TX)', 'Point Venture (TX)',
       'Mustang Ridge (TX)', 'Caldwell (TX)', 'Webberville (TX)']
```

```python
# Eliminating cities with 2 letter words as in two_word_cities since they do not give us the right addresses.
# Reduces to 105224

refined_locations = location_df[~location_df['Found_Location'].isin(short_cities_list)]
refined_locations
```

```python
uniqueId = refined_locations["Animal_ID"].unique() 
uniqueId
```

```python
refined_locations= refined_locations.drop_duplicates(subset=['Animal_ID'],keep=False)
refined_locations
```

```python
address_start_digit = refined_locations.Found_Location.str[0].str.isdigit()
address_start_digit
```

```python
refined_locations = refined_locations.loc[address_start_digit]
refined_locations
```

```python
# # unique_list = refined_locations.Found_Location.unique().tolist()
# refined_locations.Found_Location.unique().tolist()
# unique_df = refined_locations[refined_locations['Found_Location'].isin(refined_locations.Found_Location.unique().tolist())]
# unique_df


refined_locations.Found_Location.nunique()
```

```python
refined_locations["State"] = refined_locations["Found_Location"].str.split().str[-1]
```

```python
refined_locations["City"] = refined_locations["Found_Location"].str.split().str[-2]
```

```python

refined_locations['Address'] = refined_locations["Found_Location"].apply(lambda s: ', '.join([' '.join(x.split()[:-2]) for x in s.split(',')]))
refined_locations
```

```python
refined_locations = refined_locations.drop_duplicates(subset=["Found_Location"],keep="first")
refined_locations
```
